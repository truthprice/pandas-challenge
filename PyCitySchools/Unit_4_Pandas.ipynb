```python
# Dependencies
import pandas as pd
```

```python
# Load CSV files
school_data_csv = "Resources/schools_complete.csv"
student_data_csv = "Resources/students_complete.csv"
```

```python
# Read CSV files and store in dataframes
school_data_df = pd.read_csv(school_data_csv)
student_data_df = pd.read_csv(student_data_csv)
```

```python
# Combine school and student data
combined_data = pd.merge(school_data_df, student_data_df, how="left", on=["school_name", "school_name"])
combined_data.head()
```

```python
# Calculate the total number of schools
total_schools = [len(combined_data['school_name'].unique())]
# Calculate the total number of students
total_students = [len(combined_data['Student ID'].unique())]
all_students = len(combined_data['Student ID'].unique())
# Calculate the total budget
total_budget = school_data_df['budget'].sum()
# Calculate the average math score 
math_average = combined_data['math_score'].mean()
# Calculate the average reading score
reading_average = combined_data['reading_score'].mean()
# Calculate the percentage of students with a passing math score (70 or greater)
passing_math = (len(combined_data.loc[combined_data['math_score'] >= 70]))/all_students * 100
# Calculate the percentage of students with a passing reading score (70 or greater)
passing_reading = (len(combined_data.loc[combined_data['reading_score'] >= 70]))/all_students * 100
# Calculate the percentage of students who passed math and reading (% Overall Passing)
failed = student_data_df[(student_data_df['math_score'] < 70) | (student_data_df['reading_score'] < 70)].index
student_data_df.drop(failed, inplace=True)
overall_passing = (len(student_data_df['Student ID']))/all_students * 100

# Create a dataframe to hold the above results

district_summary = pd.DataFrame()
district_summary['Total Schools'] = total_schools
district_summary['Total Students'] = total_students
district_summary['Total Budget'] = total_budget
district_summary['Average Math Score'] = math_average
district_summary['Average Reading Score'] = reading_average
district_summary['% Passing Math'] = passing_math
district_summary['% Passing Reading'] = passing_reading
district_summary['% Overall Passing'] = overall_passing
district_summary
```
